```python
def map_func(x):
    judge_set = set(
        ["Carpool", 
         "SpFlash", 
         "APlusFlash", 
         "Unione", 
         "YouXiang", 
         "Flash"]
    )
    def get_product_set(product_set):
        rst = list()
        for product in product_set:
            if product != "Other":
                rst.append(product)
        return rst
    def get_time_span(time1, time2):
        from datetime import datetime 
        time1 = datetime.strptime(time1,"%Y-%m-%d %H:%M:%S")
        time2 = datetime.strptime(time2,"%Y-%m-%d %H:%M:%S")
        return (time2 - time1).seconds
    def geodistance(lng1,lat1,lng2,lat2):
        from math import radians, cos, sin, asin, sqrt
        #lng1,lat1,lng2,lat2 = (120.12802999999997,30.28708,115.86572000000001,28.7427)
        lng1, lat1, lng2, lat2 = map(radians, [float(lng1), float(lat1), float(lng2), float(lat2)]) # 经纬度转换成弧度
        dlon=lng2-lng1
        dlat=lat2-lat1
        a=sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
        distance=2*asin(sqrt(a))*6371*1000 # 地球平均半径，6371km
        return distance
    def get_re_call_order_label(h5_data, call_order_dict):
        h5_call_time = h5_data["h5_call_time"]
        h5_lat = h5_data["h5_call_lat"]
        h5_lng = h5_data["h5_call_lng"]
        h5_trace = h5_data["h5_athena_id_t1"]
        for call_trace in call_order_dict:
            call_dict = call_order_dict[call_trace]
            if len(call_dict) == 0:
                continue
            call_time = call_dict["call_time"]
            call_lat = call_dict["call_lat"]
            call_lng = call_dict["call_lng"]
            if call_trace != h5_trace:
                if get_time_span(h5_call_time, call_time) <= 300 and geodistance(h5_lng, h5_lat, call_lng, call_lat) <= 500:
                    return 0
        return 1
        
    k, vv = x
    h5_data_dict = dict()
    call_order_dict = dict()
    for v in vv:
        h5_data = dict()
        h5_athena_id_t1 = str(v["h5_athena_id_t1"])
        try:
            h5_data["h5_athena_id_t1"] = str(v["h5_athena_id_t1"])
            h5_data["h5_accept_or_not"] = int(v["h5_accept_or_not"])
            h5_data["h5_city_id"] = int(v["h5_city_id"])
            h5_data["h5_dt"] = str(v["h5_dt"])
            h5_data["h5_is_td_finish_flag"] = int(v["h5_is_td_finish_flag"])
            h5_data["h5_is_td_cancel_flag"] = int(v["h5_is_td_cancel_flag"])
            h5_data["h5_is_td_answer_flag"] = int(v["h5_is_td_answer_flag"])
            h5_data["h5_is_td_call_flag"] = int(v["h5_is_td_call_flag"])
            h5_data["h5_bubble_trace_id_base"] = str(v["h5_bubble_trace_id_base"])
            h5_data["h5_call_lng"] = float(v["h5_call_lng"])
            h5_data["h5_call_lat"] = float(v["h5_call_lat"])
            h5_data["h5_call_time"] = str(v["h5_call_time"])
            h5_data["h5_athena_id"] = str(v["h5_athena_id"])
            h5_data["h5_product_set"] = str(v["h5_product_set"])
            h5_data["h5_pl_num"] = int(v["h5_pl_num"])
            h5_data["h5_passenger_id"] = str(v["h5_passenger_id"])
            h5_data["h5_bubble_trace_id_multi"] = str(v["h5_bubble_trace_id_multi"])
        except:
            h5_data = dict()
        call_dict = dict()
        call_bubble_trace_id = str(v["call_bubble_trace_id"])
        try:
            call_dict["call_bubble_trace_id"] = str(v["call_bubble_trace_id"])
            call_dict["call_time"] = str(v["call_time"])
            call_dict["call_passenger_id"] = str(v["call_passenger_id"])
            call_dict["call_order_id"] = str(v["call_order_id"])
            call_dict["call_product_category"] = str(v["call_product_category"])
            call_dict["call_pl_num"] = int(v["call_pl_num"])
            call_dict["call_gmv_amt"] = float(v["call_gmv_amt"])
            call_dict["call_lng"] = float(v["call_lng"])
            call_dict["call_lat"] = float(v["call_lat"])
            call_dict["call_is_td_finish_flag"] = int(v["call_is_td_finish_flag"])
            call_dict["call_time_t2"] = str(v["call_time_t2"])
        except:
            call_dict = dict()
        h5_data_dict[h5_athena_id_t1] = h5_data
        call_order_dict[call_bubble_trace_id] = call_dict
    rst = list()
    for h5_trace in h5_data_dict:
        h5_data = h5_data_dict[h5_trace]
        try:
            h5_accept_or_not = h5_data["h5_accept_or_not"]
            h5_is_td_finish_flag = h5_data["h5_is_td_finish_flag"]
            h5_bubble_trace_id_base = h5_data["h5_bubble_trace_id_base"]
            h5_bubble_trace_id_multi = h5_data["h5_bubble_trace_id_multi"]
            h5_trace = h5_data["h5_athena_id_t1"]
            label = 0
            if h5_accept_or_not == 1 and h5_is_td_finish_flag == 1: #接受且完单
                label = 0
            if h5_accept_or_not == 1 and h5_is_td_finish_flag == 0: #接受且不完单
                label = get_re_call_order_label(h5_data, call_order_dict)
            if h5_accept_or_not == 0:
                label = 0
            if h5_trace is not None and h5_trace != np.nan and h5_trace != "" and (h5_bubble_trace_id_base is None or h5_bubble_trace_id_base == np.nan or h5_bubble_trace_id_base == "") and (h5_bubble_trace_id_multi is None or h5_bubble_trace_id_multi == np.nan or h5_bubble_trace_id_multi == ""):
                label = get_re_call_order_label(h5_data, call_order_dict)
            rst.append((h5_trace, [label]))
        except:
            continue
    if len(rst) == 0:
        rst = [("", [-1])]
    return rst
```

```python
import pandas as pd
def to_pandas_1(data):
    rst = pd.DataFrame([v for v in data])
    rst.columns = [
        "trace_id",
        "value"
    ]
    label_list = list()
    for i in range(len(rst["value"].values)):
        label_list.append(rst["value"].values[i][0])
    rst["label"] = label_list
    return rst[["trace_id", "label"]]
```
