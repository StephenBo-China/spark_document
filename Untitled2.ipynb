```python
def delete_item_in_l2_from_l1(l1, l2):
    return [x for x in l1 if x not in l2]

def get_multi_unavailable_list(data):
    data_tmp = data[
        (data["product_athena_key"] == "Carpool")
        | (data["product_athena_key"] == "SpFlash")
        | (data["product_athena_key"] == "APlusFlash")
        | (data["product_athena_key"] == "Unione")
        | (data["product_athena_key"] == "YouXiang")
    ]
    answer_multi_unavailable_list = data_tmp[
        data_tmp["response_rate"] < data_tmp["answer_rate"] + data_tmp["AnswerRateGap"]
    ]["product_athena_key"].values
    price_multi_unavailable_list = data_tmp[
        data_tmp["pre_price"] > data_tmp["final_fee"] * data_tmp["PriceRatio"]
    ]["product_athena_key"].values
    price_multi_unavailable_list = delete_item_in_l2_from_l1(price_multi_unavailable_list, answer_multi_unavailable_list)
    accept_multi_unavailable_list = data_tmp[
        data_tmp["premiumH5AcceptRatio"] <= data_tmp["AcceptRateThreshold"]
    ]["product_athena_key"].values
    accept_multi_unavailable_list = delete_item_in_l2_from_l1(accept_multi_unavailable_list, price_multi_unavailable_list)
    accept_multi_unavailable_list = delete_item_in_l2_from_l1(accept_multi_unavailable_list, answer_multi_unavailable_list)
    data_tmp = data[
        (data["product_athena_key"] != "Carpool")
        & (data["product_athena_key"] != "SpFlash")
        & (data["product_athena_key"] != "APlusFlash")
        & (data["product_athena_key"] != "Unione")
        & (data["product_athena_key"] != "YouXiang")
    ]
    multi_unavailable_list = np.array([])
    other_multi_unavailable_list = data_tmp[data_tmp["product_athena_key"] != "Flash"]["product_athena_key"].values
    multi_unavailable_list = np.hstack((multi_unavailable_list, other_multi_unavailable_list))
    multi_unavailable_list = np.hstack((multi_unavailable_list, answer_multi_unavailable_list))
    multi_unavailable_list = np.hstack((multi_unavailable_list, price_multi_unavailable_list))
    multi_unavailable_list = np.hstack((multi_unavailable_list, accept_multi_unavailable_list))
    return other_multi_unavailable_list, answer_multi_unavailable_list, price_multi_unavailable_list, accept_multi_unavailable_list, multi_unavailable_list
```

```python
def get_single_available_list(data):
    data_tmp = data[
        (data["product_athena_key"] == "Carpool")
        | (data["product_athena_key"] == "SpFlash")
        | (data["product_athena_key"] == "APlusFlash")
        | (data["product_athena_key"] == "Unione")
        | (data["product_athena_key"] == "YouXiang")
    ]
    single_available_list = data_tmp[
        (data_tmp["response_rate"] >= data_tmp["answer_rate"] + data_tmp["AnswerRateGap"])
        & (data_tmp["pre_price"] <= data_tmp["final_fee"] * data_tmp["PriceRatio"])
        & (data_tmp["premiumH5AcceptRatio"] > data_tmp["AcceptRateThreshold"])
    ]["product_athena_key"].values
    single_available_list = np.append(single_available_list, "Flash")
    return single_available_list
```

```python
def judge_multi_guide(multi_unavailable_list, select_product_set, answer_multi_unavailable_list, price_multi_unavailable_list, accept_multi_unavailable_list, other_multi_unavailable_list):
    guide_judge = False
    not_guide_reason = {}
    select_product_set = select_product_set.split(",")
    inner_set = list(set(multi_unavailable_list).intersection(set(select_product_set)))
    if len(inner_set) == 0:
        guide_judge = True
    else:
        guide_judge = False
        for product in inner_set:
            if product in answer_multi_unavailable_list:
                not_guide_reason[product] = "answer"
            if product in price_multi_unavailable_list:
                not_guide_reason[product] = "price"
            if product in accept_multi_unavailable_list:
                not_guide_reason[product] = "accept"
            if product in other_multi_unavailable_list:
                not_guide_reason[product] = "other"
    return guide_judge, not_guide_reason                                        
```

```python
def judge_single_guide(single_available_list, select_product_set, answer_multi_unavailable_list, price_multi_unavailable_list, accept_multi_unavailable_list, other_multi_unavailable_list):
    guide_judge = False
    not_guide_reason = {}
    select_product_set = select_product_set.split(",")
    inner_set = list(set(single_available_list).intersection(set(select_product_set)))
    if len(inner_set) == 0:
        guide_judge = False
        for product in select_product_set:
            if product in answer_multi_unavailable_list:
                not_guide_reason[product] = "answer"
            if product in price_multi_unavailable_list:
                not_guide_reason[product] = "price"
            if product in accept_multi_unavailable_list:
                not_guide_reason[product] = "accept"
            if product in other_multi_unavailable_list:
                not_guide_reason[product] = "other"
    else:
        guide_judge = True
    return guide_judge, not_guide_reason   
```

```python
def map_func(x):
    k, vv = x
    trace_id = k
    total_filter, answer_filter, price_filter, accept_filter, other_filter = 0, 0, 0, 0, 0
    for v in vv:
        product_set = v["product_set"]
        pl_num = v["pl_num"]
        premiumH5AcceptRatio = v["premiumH5AcceptRatio"]
        response_rate = v["response_rate"]
        pre_price = v["pre_price"]
        AcceptRateThreshold = v["AcceptRateThreshold"]
        PriceRatio = v["PriceRatio"]
        AnswerRateGap = v["AnswerRateGap"]
        final_fee = v["final_fee"]
        accept_rate = v["accept_rate"]
        answer_rate = v["answer_rate"]
        product_athena_key = v["product_athena_key"]
        product_judge_list = ["Carpool", "SpFlash", "APlusFlash", "Unione", "YouXiang"]
        total_judge = False
        if product_athena_key in product_judge_list:
            if response_rate < answer_rate + AnswerRateGap:
                answer_filter = answer_filter + 1
                total_judge = True
            if pre_price > final_fee + PriceRatio:
                price_filter = price_filter + 1
                total_judge = True
            if premiumH5AcceptRatio <= AcceptRateThreshold:
                accept_filter = accept_filter + 1
                total_judge = True
        else:
            other_filter = other_filter + 1
        if total_judge: 
            total_filter = total_filter + 1
    return (product_athena_key, [total_filter, answer_filter, price_filter, accept_filter, other_filter])
```
