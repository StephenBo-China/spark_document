```python
def sum_map(x):
    sum_gmv = 0.
    k, vv = x
    for v in vv:
        gmv_amt = float(v["gmv_amt"])
        sum_gmv += gmv_amt
        break
    return [("", [sum_gmv])]
def trace_num_map(x):
    sum_gmv = 0.
    k, vv = x
    total_num = 1
    for v in vv:
        gmv_amt = float(v["gmv_amt"])
        sum_gmv += gmv_amt
    return [("", [1])]
```
