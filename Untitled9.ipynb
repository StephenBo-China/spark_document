```python
def map_func(x):
    judge_set = set(
        ["Carpool", 
         "SpFlash", 
         "APlusFlash", 
         "Unione", 
         "YouXiang", 
         "Flash"]
    )
    def get_product_set(product_set):
        rst = list()
        for product in product_set:
            if product != "Other":
                rst.append(product)
        return rst
    def get_time_span(time1, time2):
        from datetime import datetime 
        time1 = datetime.strptime(time1,"%Y-%m-%d %H:%M:%S")
        time2 = datetime.strptime(time2,"%Y-%m-%d %H:%M:%S")
        return (time2 - time1).seconds
    def geodistance(lng1,lat1,lng2,lat2):
        from math import radians, cos, sin, asin, sqrt
        #lng1,lat1,lng2,lat2 = (120.12802999999997,30.28708,115.86572000000001,28.7427)
        lng1, lat1, lng2, lat2 = map(radians, [float(lng1), float(lat1), float(lng2), float(lat2)]) # 经纬度转换成弧度
        dlon=lng2-lng1
        dlat=lat2-lat1
        a=sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
        distance=2*asin(sqrt(a))*6371*1000 # 地球平均半径，6371km
        return distance
    def get_re_call_order_gmv(h5_data, call_order_dict):
        h5_call_time = h5_data["h5_call_time"]
        h5_lat = h5_data["h5_lat"]
        h5_lng = h5_data["h5_lng"]
        h5_trace = h5_data["h5_bubble_trace_id"]
        gmv = 0.0
        for call_trace in call_order_dict:
            call_dict = call_order_dict[call_trace]
            if len(call_dict) == 0:
                continue
            call_gmv = call_dict["call_order_gmv_amt"]
            call_time = call_dict["call_order_call_time"]
            call_lat = call_dict["call_order_call_lat"]
            call_lng = call_dict["call_order_call_lng"]
            if call_trace != h5_trace:
                if get_time_span(h5_call_time, call_time) <= 300 and geodistance(h5_lng, h5_lat, call_lng, call_lat) <= 500:
                    #发单时间小于等于5分钟，发单距离小于等于500米
                    return call_gmv
        return gmv                 
        
    h5_data_dict = dict()
    call_order_dict = dict()
    apollo_order_dict = dict()
    k, vv = x
    # Process and Record Data
    for v in vv:
        exp_group = str(v["exp_group"])
        apollo_key = str(v["apollo_key"])
        city = int(v["city"])
        ymd = str(v["ymd"])
        apollo_dict = dict()
        apollo_dict["exp_group"] = exp_group
        apollo_dict["apollo_key"] = apollo_key
        apollo_dict["city"] = city
        apollo_dict["ymd"] = ymd
        # Process H5 Data
        if v["h5_product_set"] is None or v["h5_product_set"] == "null" or v["h5_product_set"] == "":
            continue
        h5_gmv = float(v["h5_gmv"])
        h5_bubble_trace_id = str(v["h5_bubble_trace_id"])
        h5_lat = float(v["h5_lat"])
        h5_lng = float(v["h5_lng"])
        h5_call_time = str(v["h5_call_time"])
        h5_product_set = str(v["h5_product_set"])
        h5_pl_num = int(v["h5_pl_num"])
        h5_data = dict()
        h5_data["h5_gmv"] = h5_gmv
        h5_data["h5_lat"] = h5_lat
        h5_data["h5_lng"] = h5_lng
        h5_data["h5_call_time"] = h5_call_time
        h5_data["h5_product_set"] = h5_product_set
        h5_data["h5_bubble_trace_id"] = h5_bubble_trace_id
        h5_data_dict[h5_bubble_trace_id] = h5_data
        apollo_order_dict[h5_bubble_trace_id] = apollo_dict
        # Process Call Order Data
        call_dict = dict()
        try:
            call_order_bubble_trace_id = str(v["call_order_bubble_trace_id"])
            call_order_passenger_id = str(v["call_order_passenger_id"])
            call_order_call_time = str(v["call_order_call_time"])
            call_order_product_category = str(v["call_order_product_category"])
            call_order_order_id = str(v["call_order_order_id"])
            call_order_pl_num = int(v["call_order_pl_num"])
            call_order_gmv_amt = float(v["call_order_gmv_amt"])
            call_order_call_lat = float(v["call_order_call_lat"])
            call_order_call_lng = float(v["call_order_call_lng"])
            call_order_is_td_finish_flag = int(v["call_order_is_td_finish_flag"])
            call_dict["call_order_passenger_id"] = call_order_passenger_id
            call_dict["call_order_call_time"] = call_order_call_time
            call_dict["call_order_product_category"] = call_order_product_category
            call_dict["call_order_order_id"] = call_order_order_id
            call_dict["call_order_pl_num"] = call_order_pl_num
            call_dict["call_order_gmv_amt"] = call_order_gmv_amt
            call_dict["call_order_call_lat"] = call_order_call_lat
            call_dict["call_order_call_lng"] = call_order_call_lng
            call_dict["call_order_is_td_finish_flag"] = call_order_is_td_finish_flag
        except:
            call_dict = dict()
        call_order_dict[call_order_bubble_trace_id] = call_dict
    rst = list()
    for h5_trace in h5_data_dict:
        h5_data = h5_data_dict[h5_trace]
        h5_gmv = h5_data["h5_gmv"]
        exp_group = apollo_order_dict[h5_trace]["exp_group"]
        product_set = sorted(get_product_set(h5_data["h5_product_set"].split(",")))
        if not set(product_set).issubset(judge_set):
            break
        rst_key = ",".join([ x for x in product_set])
        if h5_gmv == 0.:
            h5_gmv = get_re_call_order_gmv(h5_data, call_order_dict)
        treat_num = 0
        control_num = 0
        treat_gmv = 0.
        control_gmv = 0.
        if exp_group == "treatment_group":
            treat_gmv = h5_gmv
            treat_num = 1
        if exp_group == "control_group":
            control_gmv = h5_gmv
            control_num = 1
        rst.append((rst_key, [treat_gmv, control_gmv, treat_num, control_num]))
    return rst
```

```python
def to_pandas_1(data):
    rst = pd.DataFrame([v for v in data])
    rst.columns = [
        "key",
        "value"
    ]
    treat_gmv_list = list()
    control_gmv_list = list()
    treat_num_list = list()
    control_num_list = list()
    for i in range(len(rst["value"].values)):
        treat_gmv_list.append(rst["value"].values[i][0])
        control_gmv_list.append(rst["value"].values[i][1])
        treat_num_list.append(rst["value"].values[i][2])
        control_num_list.append(rst["value"].values[i][3])
    rst["treat_gmv"] = treat_gmv_list
    rst["control_gmv"] = control_gmv_list
    rst["treat_num"] = treat_num_list
    rst["control_num"] = control_num_list
    return rst[["key", "treat_gmv", "control_gmv", "treat_num", "control_num"]]
```

```python
def get_gmv(rst_pd_4, no_h5_selection_list, rate_judge):
    treat_sum_gmv_list = list()
    treat_order_num = list()
    h5_num_list = list()
    rst = rst_pd_4.copy()
    for i in range(len(rst_pd_4)):
        treat_gmv = rst_pd_4["实验组贡献总GMV"].values[i]
        control_gmv = rst_pd_4["对照组贡献总GMV"].values[i]
        treat_num = rst_pd_4["treat_num"].values[i]
        control_num = rst_pd_4["control_num"].values[i]
        selection = rst_pd_4["勾选车型"].values[i]
        h5_num = rst_pd_4["H5弹窗数"].values[i]
        filter_rate = rst_pd_4["H5弹窗后直接退出或返回流失率"].values[i]
        total_filter_rate = rst_pd_4["总流失率"].values[i]
        if selection not in no_h5_selection_list:
            treat_sum_gmv_list.append(treat_gmv)
            treat_order_num.append(treat_num)
            h5_num_list.append(h5_num)
        else:
            treat_sum_gmv_list.append(control_gmv)
            treat_order_num.append(control_num)
            h5_num_list.append(0.)
    rst["实验组贡献总GMV"] = treat_sum_gmv_list
    rst["treat_num"] = treat_order_num
    rst["H5弹窗数"] = h5_num_list
    return rst
```
