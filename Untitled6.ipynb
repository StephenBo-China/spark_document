# 统计不同勾选车型alpha: 
$$
alpha = \frac{真实GMV}{\sum 每个勾选车型preprice}
$$

```python
x = data_1.head(9)
x
```

```python
def map_func_1(x):
    judge_set = set(
        ["Carpool", 
         "SpFlash", 
         "APlusFlash", 
         "Unione", 
         "YouXiang", 
         "Flash"]
    )#只回溯pk决策策略中实验组需要判定的勾选车型
    def get_product_set(product_set):
        rst = list()
        for product in product_set:
            if product != "Other":
                rst.append(product)
        return rst
    k, vv = x
    trace_id = k
    rst_key = ""
    real_gmv = 0.
    pre_gmv = 0.
    process_set = set()
    for v in vv:
        if v["product_set"] is None or v["product_set"] == "null" or v["product_set"] == "":
            break
        product_set = sorted(get_product_set(v["product_set"].split(",")))
        if not set(product_set).issubset(judge_set):
            break
        rst_key = ",".join([ x for x in product_set])
        product_athena_key = v["product_athena_key"]
        final_fee = float(v["final_fee"])
        original_fee = float(v["original_fee"])
        pre_price = float(v["pre_price"])
        gmv_amt = float(v["gmv_amt"])
        if product_athena_key in product_set and product_athena_key not in process_set:
            pre_gmv += original_fee * 1.
            process_set.add(product_athena_key)
        real_gmv = gmv_amt
    if len(process_set) != len(product_set):#保证所有勾选车型的预估价都可以取到,不能取到的数据剔除
        rst_key = ""
        real_gmv = 0.
        pre_gmv = 0.
    return [(rst_key, [real_gmv, pre_gmv, 1])]
```

```python
def reduce_func(v1, v2):
    print("RUN reduce_func")
    return list(map(sum, zip(v1, v2))) 
```

```python
rst = data_1.rdd \
    .map(lambda x: (x["traceid"], x)) \
    .groupByKey() \
    .flatMap(map_func_1) \
    .reduceByKey(reduce_func) \
    .collect()
rst
```

```python
def to_pandas_1(data):
    rst = pd.DataFrame([v for v in data])
    rst.columns = [
        "key",
        "value"
    ]
    real_gmv_list = list()
    pre_gmv_list = list()
    for i in range(len(rst["value"].values)):
        real_gmv_list.append(rst["value"].values[i][0])
        pre_gmv_list.append(rst["value"].values[i][1])
    rst["real_gmv"] = real_gmv_list
    rst["pre_gmv"] = pre_gmv_list
    rst["alpha"] = rst["real_gmv"].values / rst["pre_gmv"].values
    return rst[["key", "alpha"]]
```

```python
rst_pd = to_pandas_1(rst)
rst_pd
```

```python
rst_pd.to_csv(
    "./rst/alpha_value.csv", 
    index = False, 
    header = True, 
    sep = ",", 
    encoding = "utf-8_sig"
)
```

# 所有h5弹窗后接受了h5的冒泡
## 对比专车完单后的真实GMV，与不弹窗后得到的GMV

```python
def map_func(x, alpha_dict):
    judge_set = set(["Carpool", "SpFlash", "APlusFlash", "Unione", "YouXiang", "Flash"])
    def get_product_set(product_set):
        rst = list()
        for product in product_set:
            if product != "Other":
                rst.append(product)
        return rst
    k, vv = x
    trace_id = k
    rst_key = ""
    premium_gmv = 0.
    selection_gmv = 0.
    print("Processing trace_id=" + str(trace_id))
    process_set = set()
    product_set = list()
    for v in vv:
        if v["product_set"] is None or v["product_set"] == "null" or v["product_set"] == "":
            break
        product_set = sorted(get_product_set(v["product_set"].split(",")))
        if not set(product_set).issubset(judge_set):
            break
        rst_key = ",".join([ x for x in product_set])
        product_athena_key = v["product_athena_key"]
        final_fee = float(v["final_fee"])
        pre_price = float(v["pre_price"])
        original_fee = float(v["original_fee"])
        gmv_amt = float(v["gmv_amt"])
        if product_athena_key in product_set and product_athena_key not in process_set:
            selection_gmv += original_fee * 1.
            process_set.add(product_athena_key)
        premium_gmv = gmv_amt
    alpha = alpha_dict[rst_key]
    selection_gmv = alpha * selection_gmv
    total_num = 1
    if len(process_set) != len(product_set):
        rst_key = ""
        selection_gmv = 0.
        premium_gmv = 0.
        total_num = 0.
    return [(rst_key, [selection_gmv, premium_gmv, total_num])]
```

```python
rst_final = data_2.rdd \
    .map(lambda x: (x["traceid"], x)) \
    .groupByKey() \
    .flatMap(lambda x: map_func(x, alpha_dict)) \
    .reduceByKey(reduce_func) \
    .collect()
rst_final
```

```python
def to_pandas_2(data):
    rst = pd.DataFrame([v for v in data])
    rst.columns = [
        "key",
        "value"
    ]
    premium_gmv_list = list()
    selection_gmv_list = list()
    total_num_list = list()
    for i in range(len(rst["value"].values)):
        selection_gmv_list.append(rst["value"].values[i][0])
        premium_gmv_list.append(rst["value"].values[i][1])
        total_num_list.append(rst["value"].values[i][2])
    rst["勾选车型"] = rst["key"].values
    rst["h5发单数"] = total_num_list
    rst["勾选车型总GMV"] = selection_gmv_list
    rst["专车完单总GMV"] = premium_gmv_list
    rst["勾选车型单均GMV"]= rst["勾选车型总GMV"].values / rst["h5发单数"].values
    rst["专车完单单均GMV"]= rst["专车完单总GMV"].values / rst["h5发单数"].values
    return rst[["勾选车型", "勾选车型总GMV","专车完单总GMV", "勾选车型单均GMV", "专车完单单均GMV", "h5发单数"]]
```

```python
def map_func_3(x):
    judge_set = set(
        ["Carpool", 
         "SpFlash", 
         "APlusFlash", 
         "Unione", 
         "YouXiang", 
         "Flash"]
    )#只回溯pk决策策略中实验组需要判定的勾选车型
    def get_product_set(product_set):
        rst = list()
        for product in product_set:
            if product != "Other":
                rst.append(product)
        return rst
    k, vv = x
    trace_id = k
    rst_key = ""
    g1,g2,g3,g4,g5,g6,g7 = 0., 0., 0., 0., 0., 0., 0.
    g1_num, g2_num, g3_num, g4_num, g5_num, g6_num, g7_num = 0, 0, 0, 0, 0, 0, 0
    is_td_finish_flag, is_td_answer_flag, h5_accept_or_not, h5_show_or_not = 0, 0, 0, 0
    real_gmv = 0.
    pre_gmv = 0.
    process_set = set()
    product_set = list()
    for v in vv:
        if v["product_set"] is None or v["product_set"] == "null" or v["product_set"] == "":
            break
        product_set = sorted(get_product_set(v["product_set"].split(",")))
        if not set(product_set).issubset(judge_set):
            break
        rst_key = ",".join([ x for x in product_set])
        product_athena_key = v["product_athena_key"]
        final_fee = float(v["final_fee"])
        original_fee = float(v["original_fee"])
        pre_price = float(v["pre_price"])
        gmv_amt = float(v["gmv_amt"])
        is_td_finish_flag = int(v["is_td_finish_flag"])
        is_td_answer_flag = int(v["is_td_answer_flag"])
        h5_accept_or_not = int(v["h5_accept_or_not"])
        h5_show_or_not = int(v["h5_show_or_not"])
        if product_athena_key in product_set and product_athena_key not in process_set:
            pre_gmv += original_fee * 1.
            process_set.add(product_athena_key)
        real_gmv = gmv_amt
    if len(process_set) != len(product_set):#保证所有勾选车型的预估价都可以取到,不能取到的数据剔除
        rst_key = ""
        g1,g2,g3,g4,g5,g6,g7 = 0., 0., 0., 0., 0., 0., 0.
    else:
        if h5_show_or_not: #H5拦截
            if h5_accept_or_not: #H5发单
                if is_td_answer_flag: #H5应答
                    if is_td_finish_flag: #完单
                        g1 = real_gmv
                        g1_num = 1
                    else: #未完单
                        g2 = real_gmv
                        g2_num = 1
                else: #H5不应答
                    g3 = real_gmv
                    g3_num = 1
            else: #H5不发单
                if is_td_answer_flag: #应答
                    if is_td_finish_flag: #完单
                        g4 = real_gmv
                        g4_num = 1
                    else: #未完单
                        g5 = real_gmv
                        g5_num = 1
                else: #不应答
                    g6 = real_gmv
                    g6_num = 1
        else: #H5不拦截
            g7 = real_gmv
            g7_num = 1
    return [(rst_key, [g1, g2, g3, g4, g5, g6, g7, g1_num, g2_num, g3_num, g4_num, g5_num, g6_num, g7_num])]
```

```python
def to_pandas_2(data):
    rst = pd.DataFrame([v for v in data])
    rst.columns = [
        "key",
        "value"
    ]
    g1_list = list()
    g2_list = list()
    g3_list = list()
    g4_list = list()
    g5_list = list()
    g6_list = list()
    g7_list = list()
    g1_num_list = list()
    g2_num_list = list()
    g3_num_list = list()
    g4_num_list = list()
    g5_num_list = list()
    g6_num_list = list()
    g7_num_list = list()
    for i in range(len(rst["value"].values)):
        g1_list.append(rst["value"].values[i][0])
        g2_list.append(rst["value"].values[i][1])
        g3_list.append(rst["value"].values[i][2])
        g4_list.append(rst["value"].values[i][3])
        g5_list.append(rst["value"].values[i][4])
        g6_list.append(rst["value"].values[i][5])
        g7_list.append(rst["value"].values[i][6])
        g1_num_list.append(rst["value"].values[i][7])
        g2_num_list.append(rst["value"].values[i][8])
        g3_num_list.append(rst["value"].values[i][9])
        g4_num_list.append(rst["value"].values[i][10])
        g5_num_list.append(rst["value"].values[i][11])
        g6_num_list.append(rst["value"].values[i][12])
        g7_num_list.append(rst["value"].values[i][13])
    rst["勾选车型"] = rst["key"].values
    rst["G1"] = g1_list
    rst["G2"] = g2_list
    rst["G3"] = g3_list
    rst["G4"] = g4_list
    rst["G5"] = g5_list
    rst["G6"] = g6_list
    rst["G7"] = g7_list
    rst["G1_num"] = g1_num_list
    rst["G2_num"] = g2_num_list
    rst["G3_num"] = g3_num_list
    rst["G4_num"] = g4_num_list
    rst["G5_num"] = g5_num_list
    rst["G6_num"] = g6_num_list
    rst["G7_num"] = g7_num_list
    rst["G1单均GMV"] = rst["G1"].values / rst["G1_num"].values
    rst["G2单均GMV"] = rst["G2"].values / rst["G2_num"].values
    rst["G3单均GMV"] = rst["G3"].values / rst["G3_num"].values
    rst["G4单均GMV"] = rst["G4"].values / rst["G4_num"].values
    rst["G5单均GMV"] = rst["G5"].values / rst["G5_num"].values
    rst["G6单均GMV"] = rst["G6"].values / rst["G6_num"].values
    rst["G7单均GMV"] = rst["G7"].values / rst["G7_num"].values
    H5_Show_GMV = np.sum(
        rst["G1"].values + rst["G2"].values + rst["G3"].values + rst["G4"].values + rst["G5"].values + rst["G6"].values
    ) / np.sum(
        rst["G1_num"].values + rst["G2_num"].values + rst["G3_num"].values + rst["G4_num"].values + rst["G5_num"].values + rst["G6_num"].values
    )
    H5_Not_Show_GMV = np.sum(rst["G7"].values) / np.sum(rst["G7_num"].values)
    H5_GMV  = np.sum(
        rst["G1"].values + rst["G2"].values + rst["G3"].values + rst["G4"].values + rst["G5"].values + rst["G6"].values
    )
    H5_Num = np.sum(
        rst["G1_num"].values + rst["G2_num"].values + rst["G3_num"].values + rst["G4_num"].values + rst["G5_num"].values + rst["G6_num"].values
    )
    Not_H5_GMV = np.sum(rst["G7"].values)
    Not_H5_Num = np.sum(rst["G7_num"].values)
    return rst[
        [
            "勾选车型", 
            "G1单均GMV","G2单均GMV", "G3单均GMV", "G4单均GMV", "G5单均GMV", "G6单均GMV", "G7单均GMV"
        ]
    ], H5_Show_GMV, H5_Not_Show_GMV, H5_GMV, H5_Num, Not_H5_GMV, Not_H5_Num
```
