```python
def map_func(x):
    judge_set = set(
        ["Carpool", 
         "SpFlash", 
         "APlusFlash", 
         "Unione", 
         "YouXiang", 
         "Flash"]
    )
    def get_product_set(product_set):
        rst = list()
        for product in product_set:
            if product != "Other":
                rst.append(product)
        return rst
    def get_time_span(time1, time2):
        from datetime import datetime 
        time1 = datetime.strptime(time1,"%Y-%m-%d %H:%M:%S")
        time2 = datetime.strptime(time2,"%Y-%m-%d %H:%M:%S")
        return (time2 - time1).seconds
    def geodistance(lng1,lat1,lng2,lat2):
        from math import radians, cos, sin, asin, sqrt
        #lng1,lat1,lng2,lat2 = (120.12802999999997,30.28708,115.86572000000001,28.7427)
        lng1, lat1, lng2, lat2 = map(radians, [float(lng1), float(lat1), float(lng2), float(lat2)]) # 经纬度转换成弧度
        dlon=lng2-lng1
        dlat=lat2-lat1
        a=sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
        distance=2*asin(sqrt(a))*6371*1000 # 地球平均半径，6371km
        return distance
    def get_quit_num(h5_data, call_data_dict):
        h5_call_time = h5_data["h5_timestampp"]
        h5_lat = h5_data["h5_flat"]
        h5_lng = h5_data["h5_flng"]
        h5_trace = h5_data["h5_athena_id"]
        h5_quit_off_num = 0
        h5_quit_resend_num = 0
        for call_trace in call_data_dict:
            call_dict = call_data_dict[call_trace]
            if len(call_dict) == 0:
                continue
            call_time = call_dict["call_order_call_time"]
            call_lat = call_dict["call_order_call_lat"]
            call_lng = call_dict["call_order_call_lng"]
            if call_trace != h5_trace and h5_lat != 0. and h5_lng != 0.:
                if get_time_span(h5_call_time, call_time) <= 300 and geodistance(h5_lng, h5_lat, call_lng, call_lat) <= 500:
                    #发单时间小于等于5分钟，发单距离小于等于500米
                    return 0, 1
            if call_trace != h5_trace and h5_lat == 0. and h5_lng == 0.:
                if get_time_span(h5_call_time, call_time) <= 300:
                    return 0, 1
        return 1, 0
    h5_data_dict = dict()
    call_data_dict = dict()
    k, vv = x
    for v in vv:
        if v["h5_product_set"] is None or v["h5_product_set"] == "null" or v["h5_product_set"] == "":
            continue
        #Record H5 Data
        h5_data = dict()
        h5_data["h5_athena_id"] = str(v["h5_athena_id"])
        h5_data["h5_accept_or_not"] = int(v["h5_accept_or_not"])
        h5_data["h5_passenger_id"] = str(v["h5_passenger_id"])
        h5_data["h5_product_set"] = str(v["h5_product_set"])
        h5_data["h5_quite_or_not"] = int(v["h5_quite_or_not"])
        try:
            h5_data["h5_flng"] = float(v["h5_flng"])
        except:
            h5_data["h5_flng"] = 0.
        try:
            h5_data["h5_flat"] = float(v["h5_flat"])
        except:
            h5_data["h5_flat"] = 0.
        h5_data["h5_timestampp"] = str(v["h5_timestampp"])
        try:
            h5_data["h5_is_td_finish_flag"] = int(v["h5_is_td_finish_flag"])
        except:
            h5_data["h5_is_td_finish_flag"] = 0
        try:
            h5_data["h5_is_td_cancel_flag"] = int(v["h5_is_td_cancel_flag"])
        except:
            h5_data["h5_is_td_cancel_flag"] = 0
        try:
            h5_data["h5_is_td_answer_flag"] = int(v["h5_is_td_answer_flag"])
        except:
            h5_data["h5_is_td_cancel_flag"] = 0
        h5_data_dict[str(v["h5_athena_id"])] = h5_data
        #Record Call Data
        call_dict = dict()
        try:
            call_dict["call_order_bubble_trace_id"] = str(v["call_order_bubble_trace_id"])
            call_dict["call_order_passenger_id"] = str(v["call_order_passenger_id"])
            call_dict["call_order_call_time"] = str(v["call_order_call_time"])
            call_dict["call_order_product_category"] = str(v["call_order_product_category"])
            call_dict["call_order_order_id"] = str(v["call_order_order_id"])
            call_dict["call_order_pl_num"] = int(v["call_order_pl_num"])
            call_dict["call_order_gmv_amt"] = float(v["call_order_gmv_amt"])
            call_dict["call_order_call_lat"] = float(v["call_order_call_lat"])
            call_dict["call_order_call_lng"] = float(v["call_order_call_lng"])
            call_dict["call_order_is_td_finish_flag"] = int(v["call_order_is_td_finish_flag"])
        except: 
            call_dict = dict()
        call_data_dict[str(v["call_order_bubble_trace_id"])] = call_dict
    rst = list()
    for h5_trace in h5_data_dict:
        h5_data = h5_data_dict[h5_trace]
        product_set = sorted(get_product_set(h5_data["h5_product_set"].split(",")))
        if not set(product_set).issubset(judge_set):
            continue
        rst_key = ",".join([ x for x in product_set])
        h5_show_num = 1
        h5_call_num = 0
        h5_not_call_num = 0
        h5_quite_num = 0
        h5_call_finish_num = 0
        h5_call_not_finish_num = 0
        h5_call_not_finish_off_num = 0
        h5_call_not_finish_resend_num = 0
        h5_quite_off_num = 0
        h5_quite_resend_num = 0
        if h5_data["h5_accept_or_not"] == 1:
            h5_call_num = 1
        if h5_data["h5_accept_or_not"] == 0:
            h5_not_call_num = 1
        if h5_data["h5_quite_or_not"] == 1:
            h5_quite_num = 1
            h5_quite_off_num, h5_quite_resend_num = get_quit_num(h5_data, call_data_dict)
        if h5_data["h5_accept_or_not"] == 1 and h5_data["h5_is_td_finish_flag"] == 1:
            h5_call_finish_num = 1
        if h5_data["h5_accept_or_not"] == 1 and h5_data["h5_is_td_finish_flag"] == 0:
            h5_call_not_finish_num = 1
            h5_call_not_finish_off_num, h5_call_not_finish_resend_num = get_quit_num(h5_data, call_data_dict)
        rst.append((rst_key, [h5_show_num, h5_call_num, h5_not_call_num, h5_quite_num, h5_call_finish_num, 
                             h5_call_not_finish_num, h5_call_not_finish_off_num, h5_call_not_finish_resend_num, 
                             h5_quite_off_num, h5_quite_resend_num]))
    return rst
```

```python
def to_pandas_1(data):
    rst = pd.DataFrame([v for v in data])
    rst.columns = [
        "key",
        "value"
    ]
#     [h5_show_num, h5_call_num, h5_not_call_num, h5_quite_num, h5_call_finish_num, 
#     h5_call_not_finish_num, h5_call_not_finish_off_num, h5_call_not_finish_resend_num, 
#     h5_quite_off_num, h5_quite_resend_num]
    h5_show_num_list = list()
    h5_call_num_list = list()
    h5_not_call_num_list = list()
    h5_quite_num_list = list()
    h5_call_finish_num_list = list()
    h5_call_not_finish_num_list = list()
    h5_call_not_finish_off_num_list = list()
    h5_call_not_finish_resend_num_list = list()
    h5_quite_off_num_list = list()
    h5_quite_resend_num_list = list()
    for i in range(len(rst["value"].values)):
        h5_show_num_list.append(rst["value"].values[i][0])
        h5_call_num_list.append(rst["value"].values[i][1])
        h5_not_call_num_list.append(rst["value"].values[i][2])
        h5_quite_num_list.append(rst["value"].values[i][3])
        h5_call_finish_num_list.append(rst["value"].values[i][4])
        h5_call_not_finish_num_list.append(rst["value"].values[i][5])
        h5_call_not_finish_off_num_list.append(rst["value"].values[i][6])
        h5_call_not_finish_resend_num_list.append(rst["value"].values[i][7])
        h5_quite_off_num_list.append(rst["value"].values[i][8])
        h5_quite_resend_num_list.append(rst["value"].values[i][9])
    rst["h5_show_num"] = h5_show_num_list
    rst["h5_call_num"] = h5_call_num_list
    rst["h5_not_call_num"] = h5_not_call_num_list
    rst["h5_quite_num"] = h5_quite_num_list
    rst["h5_call_finish_num"] = h5_call_finish_num_list
    rst["h5_call_not_finish_num"] = h5_call_not_finish_num_list
    rst["h5_call_not_finish_off_num"] = h5_call_not_finish_off_num_list
    rst["h5_call_not_finish_resend_num"] = h5_call_not_finish_resend_num_list
    rst["h5_quite_off_num"] = h5_quite_off_num_list
    rst["h5_quite_resend_num"] = h5_quite_resend_num_list
    return rst[["key", "h5_show_num", "h5_call_num", "h5_not_call_num", "h5_quite_num", 
               "h5_call_finish_num", "h5_call_not_finish_num", "h5_call_not_finish_off_num", "h5_call_not_finish_resend_num", 
                "h5_quite_off_num", "h5_quite_resend_num"]]
```
